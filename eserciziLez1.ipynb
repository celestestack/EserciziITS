## Esercizio 1: Divisione

**Consegna:** Creare un programma che permetta all'utente di effettuare la divisione tra due numeri e mostri a schermo quoziente e resto senza utilizzare né l'operatore / né l'operatore %

````python
def divisione(n1, n2):
    result = 0
    while n1 >= n2:
        n1 -= n2
        result += 1
    return result, n1


print(divisione(35, 7))
````

## Esercizio 2: Funzioni Personalizzate

**Consegna:** Implementare delle funzioni che consentano di trovare il massimo, minimo, contare il numero di elementi su una lista e ridefinire la funzione range senza usare le funzioni predefinite di python (len, max, min, range).

````python
def massimo(lista):
    '''
    funzione che individua il massimo senza usare funzione max (incredibile)
    '''
    max = lista[0]
    for i in lista:
        if i > max:
            max = i
    return max

def minimo(lista):
    '''
    funzione che individua il minimo senza usare funzione min (pazzesca)
    '''
    min = lista[0]
    for i in lista:
        if i < min:
            min = i
    return min

def conta(lista):
    contaElem = 0
    for i in lista:
        contaElem += 1
    return contaElem

def myRange(start, stop, step = 1):
    risultato = []
    corrente = start
    while corrente < stop:
        risultato.append(corrente)
        corrente += step
    return risultato

print(massimo([1, 2, 3, 4]))
print(minimo([1, 2, 3, 4]))
print(conta([1, 2, 3, 4]))
print(myRange(0, 10))
````

## Esercizio 3: Lista Booleana

**Consegna:** Creare una funzione che, data una lista di numeri ed un numero da cercare, restituisca il valore booleano True se tale numero è presente nella lista, False altrimenti

````python
listaNum = [1, 2, 3, 4, 5, 6, 7, 8, 9]
n = 7

def trovaNum(lista, num):
    listaStringata, num = str(lista), str(num)
    if num in listaStringata:
        return True
    else:
        return False
    
print(trovaNum(listaNum, n))


def trovaNum_4(lista, num):
    for i in lista:
        if i == num:
            return True
    return False
    
print(trovaNum_4(listaNum, n))
````

## Esercizio 4: Moltiplicazione

**Consegna:** Creare un programma che permetta all'utente di effettuare la moltiplicazione tra due numeri senza utilizzare l'operatore * ma calcolando il risultato attraverso somme ripetute (numeri positivi).

````python
def moltiplicazione(n1, n2):
    result = 0
    for i in range(n2):
        result += n1
    return result


print(moltiplicazione(7, 5))
````

## Esercizio 5: Occorrenze

**Consegna:** Creare una funzione che abbia due parametri: una stringa e un carattere. La funzione dovrà scorrere tutte le lettere della stringa utilizzando l'istruzione for e restituire il numero di occorrenze del carattere nella stringa.

````python
s = "ciaociaociao"

def occorrenze(stringa, carattere):
    conta_carattere = 0
    for lettera in stringa:
        if lettera == carattere:
            conta_carattere += 1
    return conta_carattere


print(occorrenze(s, 'c'))
````

## Esercizio 6: Stringa Pari

**Consegna:** Creare un programma che chieda all'utente di inserire una stringa di lunghezza pari. Se l'utente inserisce una stringa dispari, il programma dovrà chiedergli di reinserirla fino a che non inserisca una stringa di lunghezza pari.

````python
def stringaPari(stringaInput):
    lunghezza = len(stringaInput)
    if lunghezza%2 != 0:
        while True:
            strPari = input("inserire una stringa pari!")
            break
    else:
        strPari = stringaInput
    return strPari


print(stringaPari("ciao come va!"))
````

## Esercizio 7: Stringa Alternata

**Consegna:** Creare un programma che chieda all'utente di inserire un numero n e stampi una stringa lunga n caratteri dove i caratteri che la compongono saranno i caratteri @ e # alternati partendo con un carattere @

Ad esempio se l'utente dice che la stringa dovrà essere lunga 5 caratteri la stringa stampata dovrà essere: @#@#@

````python
def stringaFinale(n):
    stringaF = ""
    for i in range(n):
        if i%2 == 0:
            stringaF += "@"
        else:
            stringaF += "#"

    return stringaF

numero = int(input("inserire un valore: "))
print(stringaFinale(numero))
````

## Esercizio 9: Palindroma

**Consegna:** Scrivere una funzione che dica se una parola è palindroma ("Anna", "SOS")

````python
def palindroma(stringa):
    n = len(stringa)
    for i in range(n // 2):
        if stringa[i] != stringa[n - 1 - i]:
            return "non palindroma"
    return "palindroma"
````

## Esercizio 10: Numero Primo

**Consegna:** Scrivere una funzione che dica se un numero è primo o no

````python
def primo(numero):
    if numero < 2:
        return "non primo"
    for i in range(2, int(numero ** 0.5) + 1):
        if numero % i == 0:
            return "non primo"
    return "primo"
````

## Esercizio 12: Fibonacci

**Consegna:** Scrivere una funzione che produca una lista dei numeri di Fibonacci fino ad n

````python
def fibonacci(n):
    if n < 0:
        return "n deve essere non negativo"
    
    risultato = []
    a, b = 0, 1
    
    while a <= n:
        risultato.append(a)
        a, b = b, a + b
    
    return risultato
````

## Esercizio 13: Media

**Consegna:** Scrivere una funzione che restituisce la media dei numeri passati come lista

````python
def media(lista):    
    if len(lista) == 0:
        return "lista vuota, impossibile calcolare la media"
    
    somma = sum(lista)
    media_calcolata = somma / len(lista)
    
    return media_calcolata
````
