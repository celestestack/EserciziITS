### 1. Dato un numero scrivere se è positivo, negativo o nullo

```python
numero = float(input("Inserisci un numero: "))

if numero > 0:
    print("Il numero è positivo")
elif numero < 0:
    print("Il numero è negativo")
else:
    print("Il numero è nullo")

print("Test: 0 -> nullo")
print("Test: -3 -> negativo")
print("Test: 5 -> positivo")
```

### 2. Dati due numeri, scrivere il maggiore (verificare anche se sono uguali)

```python
num1 = float(input("Inserisci il primo numero: "))
num2 = float(input("Inserisci il secondo numero: "))

if num1 > num2:
    print(f"Il maggiore è {num1}")
elif num2 > num1:
    print(f"Il maggiore è {num2}")
else:
    print("I due numeri sono uguali")

print("Test: 4 e 4 -> I due numeri sono uguali")
print("Test: 5 e 3 -> Il maggiore è 5")
```

### 3. Dati due numeri scrivere il loro quoziente, se il divisore è diverso da 0 allora stampare la divisione altrimenti "impossibile" se il divisore = 0

```python
dividendo = float(input("Inserisci il dividendo: "))
divisore = float(input("Inserisci il divisore: "))

if divisore != 0:
    quoziente = dividendo / divisore
    print(f"Il quoziente è: {quoziente}")
else:
    print("Impossibile: divisione per zero!")
```

### 4. Data la base e l'altezza di un triangolo, scrivere l'area

```python
base = float(input("Inserisci la base del triangolo: "))
altezza = float(input("Inserisci l'altezza del triangolo: "))

area = (base * altezza) / 2
print(f"L'area del triangolo è: {area}")
```

### 9. Inserire n numeri diversi da 0 (0 per finire), contare quanti sono i numeri inseriti

```python
count = 0

print("Inserisci numeri (0 per terminare):")
while True:
    numero = float(input("Numero: "))
    if numero == 0:
        break
    count += 1

print(f"Hai inserito {count} numeri")
```

### 10. Implementare una calcolatrice chiedendo all'utente di inserire i valori e l'operatore

```python
num1 = float(input("Inserisci il primo numero: "))
operatore = input("Inserisci l'operatore (+, -, *, /): ")
num2 = float(input("Inserisci il secondo numero: "))

if operatore == "+":
    risultato = num1 + num2
elif operatore == "-":
    risultato = num1 - num2
elif operatore == "*":
    risultato = num1 * num2
elif operatore == "/":
    if num2 != 0:
        risultato = num1 / num2
    else:
        print("Errore: divisione per zero!")
        risultato = None
else:
    print("Operatore non valido!")
    risultato = None

if risultato is not None:
    print(f"Risultato: {num1} {operatore} {num2} = {risultato}")
```

### 11. Scrivete un programma che chieda all'utente di indovinare una password, ma che dia al giocatore solamente 3 possibilità, fallite le quali il programma terminerà, stampando "È troppo complicato per voi"

```python
password_segreta = "python123"
tentativi = 3

for tentativo in range(tentativi):
    password_inserita = input(f"Tentativo {tentativo + 1}/3 - Inserisci la password: ")
    
    if password_inserita == password_segreta:
        print("Password corretta! Accesso consentito.")
        break
    else:
        if tentativo < tentativi - 1:
            print("Password errata. Riprova.")
        else:
            print("È troppo complicato per voi")
```

### 14. Riscrivete il programma 'area.py', definendo funzioni separate per l'area del quadrato, del rettangolo e del cerchio (3.14 * raggio**2). Il programma deve includere anche un'interfaccia a menu

```python
import math

def area_quadrato(lato):
    return lato ** 2

def area_rettangolo(base, altezza):
    return base * altezza

def area_cerchio(raggio):
    return 3.14 * raggio ** 2

def menu_aree():
    while True:
        print("\n=== CALCOLATORE DI AREE ===")
        print("1. Area del quadrato")
        print("2. Area del rettangolo")
        print("3. Area del cerchio")
        print("4. Esci")
        
        scelta = input("Scegli un'opzione (1-4): ")
        
        if scelta == "1":
            lato = float(input("Inserisci il lato del quadrato: "))
            area = area_quadrato(lato)
            print(f"L'area del quadrato è: {area}")
            
        elif scelta == "2":
            base = float(input("Inserisci la base del rettangolo: "))
            altezza = float(input("Inserisci l'altezza del rettangolo: "))
            area = area_rettangolo(base, altezza)
            print(f"L'area del rettangolo è: {area}")
            
        elif scelta == "3":
            raggio = float(input("Inserisci il raggio del cerchio: "))
            area = area_cerchio(raggio)
            print(f"L'area del cerchio è: {area}")
            
        elif scelta == "4":
            print("Arrivederci!")
            break
            
        else:
            print("Opzione non valida!")

menu_aree()
```

### 15. Scrivere una funzione che dica se una parola è palindroma ("Anna", "SOS")

```python
def is_palindroma(parola):
    parola_pulita = parola.lower().replace(" ", "")
    return parola_pulita == parola_pulita[::-1]

print(is_palindroma("Anna"))
print(is_palindroma("ciao"))
```

### 16. Scrivere un programma Lunghezze che chiede all'utente di inserire una sequenza di stringhe e conclusa dalla stringa vuota, e poi stampa la somma delle lunghezze delle stringhe che iniziano con una lettera maiuscola. Per esempio, se si immettono le stringhe "Albero", "foglia", "Radici", "Ramo", "fiore" (e poi "" per finire), il programma stampa 16.

```python
def calcola_lunghezze_maiuscole(stringhe):
    somma = 0
    for stringa in stringhe:
        if stringa and stringa[0].isupper():
            somma += len(stringa)
    return somma

stringhe_test = ["Albero", "foglia", "Radici", "Ramo", "fiore"]
print(calcola_lunghezze_maiuscole(stringhe_test))
```
