# Esercizio: Somma Lunghezze Stringhe

Scrivere un programma che stampi la somma totale delle lunghezza delle stringhe fornite dall'utente, finché l'utente non inserisce la stringa 'exit'

```python
somma_totale = 0

while True:
    stringa = input("Inserisci una stringa: ")
    if stringa == 'exit':
        break
    somma_totale += len(stringa)

print(f"Somma totale delle lunghezze: {somma_totale}")
```

# Esercizio 2: Contare Occorrenze in una Lista

Definire una funzione che prenda due parametri, una lista ed un elemento da cercare, questa funzione consente di contare quante volte l'elemento è presente nella lista

```python
def conta_occorrenze(lista, elemento):
    contatore = 0
    for item in lista:
        if item == elemento:
            contatore += 1
    return contatore

test_lista = [1, 2, 3, 2, 4, 2, 5, 2]
elemento_da_cercare = 2
risultato = conta_occorrenze(test_lista, elemento_da_cercare)
print(f"Occorrenze di {elemento_da_cercare}: {risultato}")
```

# Esercizio 3: Somma dei Primi 500 Numeri Naturali

Scrivere il codice python che consenta di calcolare la somma dei primi 500 numeri naturali (da 0 a 500 escluso)

```python
somma = 0
i = 0
while i < 500:
    somma += i
    i += 1
print(f"La somma dei primi 500 numeri naturali è: {somma}")
```

# Esercizio 4: Contare le Vocali

Creare una funzione che restituisca quante vocali ha una stringa che viene passato come argomento

```python
def conta_vocali(stringa):
    vocali = "aeiouAEIOU"
    return sum(1 for carattere in stringa if carattere in vocali)

test_stringa = "Ciao mondo Python"
risultato = conta_vocali(test_stringa)
print(f"Vocali in '{test_stringa}': {risultato}")
```
