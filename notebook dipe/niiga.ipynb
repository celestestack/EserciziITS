## Esercizi Base

### 1. Dato un numero scrivere se è positivo, negativo o nullo

```python
# Esercizio 1: Verificare se un numero è positivo, negativo o nullo
numero = float(input("Inserisci un numero: "))

if numero > 0:
    print("Il numero è positivo")
elif numero < 0:
    print("Il numero è negativo")
else:
    print("Il numero è nullo")
```

```python
# Test dell'Esercizio 1 con valori predefiniti
def test_positivo_negativo_nullo(numero):
    if numero > 0:
        return "Il numero è positivo"
    elif numero < 0:
        return "Il numero è negativo"
    else:
        return "Il numero è nullo"

# Test con diversi valori
test_values = [5, -3, 0, 2.5, -1.2]
for val in test_values:
    risultato = test_positivo_negativo_nullo(val)
    print(f"Numero: {val} -> {risultato}")
```

### 2. Dati due numeri, scrivere il maggiore (verificare anche se sono uguali)

```python
# Esercizio 2: Trovare il maggiore tra due numeri
num1 = float(input("Inserisci il primo numero: "))
num2 = float(input("Inserisci il secondo numero: "))

if num1 > num2:
    print(f"Il maggiore è {num1}")
elif num2 > num1:
    print(f"Il maggiore è {num2}")
else:
    print("I due numeri sono uguali")
```

```python
# Test dell'Esercizio 2 con valori predefiniti
def test_maggiore(num1, num2):
    if num1 > num2:
        return f"Il maggiore è {num1}"
    elif num2 > num1:
        return f"Il maggiore è {num2}"
    else:
        return "I due numeri sono uguali"

# Test con diversi valori
test_pairs = [(5, 3), (2, 8), (4, 4), (-1, -5), (0, 0)]
for n1, n2 in test_pairs:
    risultato = test_maggiore(n1, n2)
    print(f"Confronto {n1} e {n2}: {risultato}")
```

### 3. Dati due numeri scrivere il loro quoziente, se il divisore è diverso da 0 allora stampare la divisione altrimenti "impossibile" se il divisore = 0

```python
# Esercizio 3: Calcolare il quoziente di due numeri
dividendo = float(input("Inserisci il dividendo: "))
divisore = float(input("Inserisci il divisore: "))

if divisore != 0:
    quoziente = dividendo / divisore
    print(f"Il quoziente è: {quoziente}")
else:
    print("Impossibile: divisione per zero!")
```

### 4. Data la base e l'altezza di un triangolo, scrivere l'area

```python
# Esercizio 4: Calcolare l'area di un triangolo
base = float(input("Inserisci la base del triangolo: "))
altezza = float(input("Inserisci l'altezza del triangolo: "))

area = (base * altezza) / 2
print(f"L'area del triangolo è: {area}")
```

```python
# Esercizio 10: Calcolatrice semplice
num1 = float(input("Inserisci il primo numero: "))
operatore = input("Inserisci l'operatore (+, -, *, /): ")
num2 = float(input("Inserisci il secondo numero: "))

if operatore == "+":
    risultato = num1 + num2
elif operatore == "-":
    risultato = num1 - num2
elif operatore == "*":
    risultato = num1 * num2
elif operatore == "/":
    if num2 != 0:
        risultato = num1 / num2
    else:
        print("Errore: divisione per zero!")
        risultato = None
else:
    print("Operatore non valido!")
    risultato = None

if risultato is not None:
    print(f"Risultato: {num1} {operatore} {num2} = {risultato}")
```

```python
# Esercizio 14: Calcolatore di aree con menu
import math

def area_quadrato(lato):
    """Calcola l'area del quadrato"""
    return lato ** 2

def area_rettangolo(base, altezza):
    """Calcola l'area del rettangolo"""
    return base * altezza

def area_cerchio(raggio):
    """Calcola l'area del cerchio"""
    return 3.14 * raggio ** 2

def menu_aree():
    """Menu per calcolare diverse aree"""
    while True:
        print("\n=== CALCOLATORE DI AREE ===")
        print("1. Area del quadrato")
        print("2. Area del rettangolo")
        print("3. Area del cerchio")
        print("4. Esci")
        
        scelta = input("Scegli un'opzione (1-4): ")
        
        if scelta == "1":
            lato = float(input("Inserisci il lato del quadrato: "))
            area = area_quadrato(lato)
            print(f"L'area del quadrato è: {area}")
            
        elif scelta == "2":
            base = float(input("Inserisci la base del rettangolo: "))
            altezza = float(input("Inserisci l'altezza del rettangolo: "))
            area = area_rettangolo(base, altezza)
            print(f"L'area del rettangolo è: {area}")
            
        elif scelta == "3":
            raggio = float(input("Inserisci il raggio del cerchio: "))
            area = area_cerchio(raggio)
            print(f"L'area del cerchio è: {area}")
            
        elif scelta == "4":
            print("Arrivederci!")
            break
            
        else:
            print("Opzione non valida!")

# Avvia il menu
menu_aree()
```

```python
# Esercizio 15: Verificare se una parola è palindroma
def is_palindroma(parola):
    """Verifica se una parola è palindroma"""
    # Converte in minuscolo per confronto case-insensitive
    parola_pulita = parola.lower().replace(" ", "")
    
    # Confronta la parola con la sua versione invertita
    return parola_pulita == parola_pulita[::-1]

# Test della funzione
parole_test = ["Anna", "SOS", "radar", "ciao", "osso", "A man a plan a canal Panama"]

for parola in parole_test:
    if is_palindroma(parola):
        print(f"'{parola}' è palindroma")
    else:
        print(f"'{parola}' non è palindroma")

# Test interattivo
print("\n--- Test interattivo ---")
parola_utente = input("Inserisci una parola da verificare: ")
if is_palindroma(parola_utente):
    print(f"'{parola_utente}' è palindroma!")
else:
    print(f"'{parola_utente}' non è palindroma.")
```

```python
# Esercizio 16: Programma Lunghezze
def programma_lunghezze():
    """Calcola la somma delle lunghezze delle stringhe che iniziano con maiuscola"""
    somma_lunghezze = 0
    stringhe_maiuscole = []
    
    print("Inserisci stringhe (stringa vuota per terminare):")
    
    while True:
        stringa = input("Stringa: ")
        
        # Se la stringa è vuota, termina l'inserimento
        if stringa == "":
            break
            
        # Se la stringa inizia con maiuscola, aggiungila alla somma
        if stringa[0].isupper():
            somma_lunghezze += len(stringa)
            stringhe_maiuscole.append(stringa)
    
    print(f"\nStringhe che iniziano con maiuscola: {stringhe_maiuscole}")
    print(f"Somma delle lunghezze: {somma_lunghezze}")
    
    return somma_lunghezze

# Esecuzione del programma
risultato = programma_lunghezze()

# Test con esempio fornito
print("\n--- Test con esempio fornito ---")
stringhe_esempio = ["Albero", "foglia", "Radici", "Ramo", "fiore"]
somma_esempio = 0
stringhe_maiuscole_esempio = []

for stringa in stringhe_esempio:
    if stringa[0].isupper():
        somma_esempio += len(stringa)
        stringhe_maiuscole_esempio.append(stringa)
        
print(f"Stringhe esempio: {stringhe_esempio}")
print(f"Stringhe con maiuscola: {stringhe_maiuscole_esempio}")
print(f"Lunghezze: {[len(s) for s in stringhe_maiuscole_esempio]}")
print(f"Somma: {somma_esempio} (dovrebbe essere 16)")
```
